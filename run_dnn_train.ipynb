# Ideas

- Plot the distribution of final accuracy values
- Plot this distribution at 50, 100, 150, 180, 190 and 200 epochs
- Plot a much larger ensemble, 100 + runs
- Problem with this will be the large number of lines on the figure
- By plotting the distribution at 50, 100, 150, etc... epochs, hope to be able to detect individual runs which jump around
- In more detail: Some runs are not smooth, but jump up in accuracy after some initial training time. The path the gradient descent is "unusual"
- If this unusual path jump occurs near the 200 epoch, cannot say that training for that run has reached the maximum

# Run again, but for 400 epochs instead of 200

- Is the distribution at 200 the same?
- How does the distribution change by 400?
- Is there an indication that there would be convergence after long enough training?

```python
epochs = 400

for dnn_size in range(14, 14 + 1, 1):

    # method to interrupt process
    if os.path.isfile('STOP'):
        os.remove('STOP')
        break

    # data output filename
    filename = f'dnn_accuracy_loss_dnn_size_{dnn_size}_epoch_{epochs}.csv'

    # run for range of iterations
    for i in range(1, 1000):
        run_dnn_train.run_dnn_train(filename, dnn_size, i, epochs)

# matplotlib figure filename
figure_filename = f'dnn_accuracy_loss_dnn_size_{dnn_size}_epoch_{epochs}.png'

run_dnn_train.plot_accuracy_loss(filename, figure_filename)

for epoch in [200, 400]:
    histogram_figure_filename = f'dnn_accuracy_distribution_epoch_{epoch}.png'
    run_dnn_train.plot_accuracy_histogram(filename, histogram_figure_filename)
```

```python
# re-run plot

dnn_size = 14
epochs = 400

# data output filename
filename = f'./accuracy_histogram-3/dnn_accuracy_loss_dnn_size_{dnn_size}_epoch_{epochs}.csv'

for epoch in [200, 400]:
    histogram_figure_filename = f'./accuracy_histogram-3/dnn_accuracy_distribution_epoch_{epoch}.png'
    run_dnn_train.plot_accuracy_histogram(filename, histogram_figure_filename, epoch)
```

This shows that between 200 and 400 epochs, the DNN does not converge much more than it did from 0 to 200 epochs
